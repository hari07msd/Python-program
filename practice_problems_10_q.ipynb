```python
# Q1. Check in which quadrant a point (x, y) lies
x=int(input())
y=int(input())
if x > 0 and y > 0:
  print("First Quadrant")
elif x < 0 and y > 0:
  print("Second Quadrant")
elif x < 0 and y < 0:
  print("Third Quadrant")
elif x > 0 and y < 0:
  print("Fourth Quadrant")
elif x == 0 and y == 0:
  print("On Axis")
elif x == 0:
  print("Y axis")
else:
  print("X axis")
```

```python
# Q3. Check valid year and month and print number of days
year = int(input())
month = int(input())

if month < 1 or month > 12:
  print("Invaild")
elif month == 2 :
  if (year % 4 == 0 and year % 100 != 0):
    print("29 days")
  else:
    print("28 days")
elif month in [ 4,6,9,11]:
  print("30 days")
else:
  print("31 days")
```

```python
# Q4. Find student grade
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
# Q5. Electricity Bill
units = int(input("Enter units: "))

bill = 0

if units <= 100:
    bill = 0

elif units <= 300:
    bill = (units - 100) * 5

else:
    bill = (200 * 5) + (units - 300) * 10

if units > 500:
    bill = bill + bill * 0.10

print("Electricity Bill =", bill)
```

```python
# Q6. Product of digits
num = int(input("Enter number: "))

product = 1

while num > 0:
    digit = num % 10
    product = product * digit
    num = num // 10

print("Product =", product)
```

```python
# Q7. Sum of all factors
num = int(input("Enter number: "))

sum_factors = 0

for i in range(1, num + 1):
    if num % i == 0:
        sum_factors += i

print("Sum of factors =", sum_factors)
```

```python
# Q8. Decimal to Binary
num = int(input("Enter decimal number: "))

binary = ""

if num == 0:
    print("0")
else:
    while num > 0:
        binary = str(num % 2) + binary
        num = num // 2

    print("Binary =", binary)
```

```python
# Q9. Largest digit in a number
num = int(input("Enter number: "))

largest = 0

while num > 0:
    digit = num % 10

    if digit > largest:
        largest = digit

    num = num // 10

print("Largest digit =", largest)
```

```python
# Q10. Print all prime numbers from 1 to N
n = int(input("Enter N: "))

for num in range(2, n + 1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```
