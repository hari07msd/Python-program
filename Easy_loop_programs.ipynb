```python
# 6 count digits in number
n=int(input())
count=0
if n ==0:
  count = 1
else:
  count = 0

while n>0:
  count +=1
  n = n// 10

print(count)
```

```python
# 7 reverse a number
n=int(input())
if n==0:
  print(0)
else:
  res=0
  while n!=0:
    res = res*10+n%10
    n//=10
  print(res)
```

```python
# 8 sum of digits
n=int(input())
sum=0
while n > 0:
  i= n % 10
  sum += i
  n = n//10
print(sum)
```

```python
# 12 check prime
n=int(input())
count=0
for i in range(1,n+1):
  if n % i ==0:
    count+=1
if count==2:
  print("Prime")
else:
  print("Not Prime")
```
