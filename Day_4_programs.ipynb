```python
# Q1. Print the first digit of a given number
n=int(input())
while n>= 10:
  n= n // 10
print("First digit:",n)
```

```python
# Q2. Check whether the first digit is odd or even
n=int(input())
while n > 10:
  n = n // 10
if n % 2 == 0:
  print("Even")
else:
  print("Odd")
```

```python
# Q3. Check whether a given digit occurs in a number
n = input("Enter the number")
d = input("Enter the digit")

if d in n:
  print("The digit occurs in the number")
else:
  print("The digit does not occur in the number")
```

```python
# Q4. Count the occurrence of a digit
n=input("Enter the number")
d=input("Enter the digit")

count=0

for i in n:
  if i == d:
    count +=1
print("Count",count)
```

```python
# Q5. Check whether digits are in ascending order
n= input("Enter the number")
if list(n) == sorted(n):
  print("Ascending order")
else:
  print("Not in ascending")
```

```python
# Q6. Check whether digits are in descending order
n= input("Enter the number")
if list(n) == sorted(n, reverse=True):
  print("descending order")
else:
  print("Not in descending")
```

```python
# Q7. Swap first and last digit
n =input("Enter the number")
if len(n) == 1:
  print(n)
else:
  new = n[-1] + n[1:-1] + n[0]
  print("After swap",new)
```

```python
# Q8. Find the frequency of each digit
n =input("Enter the number")
for i in "0123456789":
  if i in n:
    print(i, "=" , n.count(i))
```
